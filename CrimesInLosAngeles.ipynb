#Crimes in Los Angeles


This project aims to analyse crimes in Los Angeles.

Source: https://www.kaggle.com/datasets/cityofLA/crime-in-los-angeles?select=Crime_Data_2010_2017.csv

```python
### Importing basic libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
filepath = "/content/drive/MyDrive/CrimeAnalysis/CrimeinLosAngeles/Crime_Data_2010_2017.csv"

df = pd.read_csv(filepath)
df.head(10)
```

```python
df = df.drop(['Crime Code 1', 'Crime Code 2', 'Crime Code 3', 'Crime Code 4', 'Crime Code Description', 'DR Number', 'MO Codes', 'Date Occurred', 'Address'], axis =1)
df.head()
```

```python
df = df.drop(['Area Name', 'Premise Description', 'Status Code', 'Status Description', 'Weapon Description', 'Cross Street', 'Location '], axis = 1)
```

```python
df.head()
```

## Data cleaning
Date reported needs to be turned into a numeric value for analysis.
Columns with missing values such as Victim Age need to get their values imputated.

```python
def dateToNumeric(date):
  date_list = date.split("/")
  date_str = ""
  for i in range(len(date_list)):
    date_str = date_str + date_list[i]
  return int(date_str)

df['Date Reported'] = df['Date Reported'].apply(dateToNumeric)
```

```python
df.head()
```

```python
def reduceTime(time):
  return int(time/100)

df['Time Occurred'] = df['Time Occurred'].apply(reduceTime)
df.head()
```

```python
def encodeGender(gender):
  if gender == 'F':
    return 0
  elif gender == 'M':
    return 1
  else:
    return None

df['Victim Sex'] = df['Victim Sex'].apply(encodeGender)
```

```python
df['Victim Age'] = df['Victim Age'].fillna(df['Victim Age'].median())
df['Victim Sex'] = df['Victim Sex'].fillna(df['Victim Sex'].median())
df['Premise Code'] = df['Premise Code'].fillna(101.0)
df['Weapon Used Code'] = df['Weapon Used Code'].fillna(0.00)
```

```python
uniqueDescents = set()
desc = df.loc[:,'Victim Descent']
for i in range(len(desc)):
  letter = str(desc[i])
  uniqueDescents.add(letter[0])
print("Number of unique districts in LA: ", len(uniqueDescents))
print("Names of the unique districts in LA: ", uniqueDescents)
```

```python
descentMap = {
    'O' : 0,
    'W' : 1,
    'X' : 2,
    'V' : 3,
    'B' : 4,
    'L' : 5,
    'P' : 6,
    'nan': 'nan',
    '-' : 'nan',
    'K' : 7,
    'Z' : 8,
    'C' : 9,
    'S' : 10,
    'U' : 11,
    'D' : 12,
    'G' : 13,
    'J' : 14,
    'I' : 15,
    'A' : 16,
    'F' : 17,
    'H' : 18
}

df['Victim Descent'] = df['Victim Descent'].map(descentMap)
```

```python
df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

X = df[['Date Reported', 'Time Occurred', 'Area ID', 'Reporting District', 'Crime Code', 'Victim Age', 'Premise Code', 'Weapon Used Code']]
y = df[['Victim Sex', 'Victim Descent']]

imputer = IterativeImputer(max_iter=10, random_state=0,estimator=LinearRegression())
X_imputed = imputer.fit_transform(df)

# Assign the imputed values back to the original DataFrame
df[['Date Reported', 'Time Occurred', 'Area ID', 'Reporting District', 'Crime Code', 'Victime Age', 'Victim Sex', 'Victim Descent', 'Premise Code', 'Weapon Used Code']] = X_imputed
df.head()
```

```python
df.head()
```

```python
df['Victim Descent'] = df['Victim Descent'].apply(np.floor)
df.head()
```

##Normalising time values
Since the values of some features are larger than the rest, scaling is done on the whole dataframe to obtain standard values.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns, index = df.index)
```

```python
df.head()
```

##Training the model

A decision tree is trained.

The dataset is split into training, testing and validation sets.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score as r2
from sklearn.model_selection import train_test_split
```

```python
X = df[['Date Reported', 'Time Occurred', 'Area ID', 'Reporting District', 'Crime Code', 'Victim Age', 'Premise Code', 'Weapon Used Code', 'Victim Age', 'Victim Sex']]
y = df['Reporting District']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle = True)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.1, shuffle = True)
```

```python
dtr = DecisionTreeRegressor(max_depth = 150, min_samples_split= 8, min_samples_leaf = 5, max_features = 10)

dtr_model = dtr.fit(X_train, y_train)
```

```python
dtr_val_pred = dtr.predict(X_val)


print("Decision Tree Regression mean squared error: ", mse(y_val, dtr_val_pred))
print("Decision Tree Regression r^2 error: ", r2(y_val, dtr_val_pred))
```

```python
dtr_test_pred = dtr.predict(X_test)


print("Decision Tree Regression mean squared error: ", mse(y_test, dtr_test_pred))
print("Decision Tree Regression r^2 error: ", r2(y_test, dtr_test_pred))
```
